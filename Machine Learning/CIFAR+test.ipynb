```python
# Actually this code was recommended in tf version 1.4.1 
# But I'm trying to use latest version. (1.8.0)

import tensorflow as tf

# check the version of tensorflow
print ("Tensorflow Version is : " + tf.__version__)

def _int64_feature(value):
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))

def _bytes_feature(value):
    return tf.train.Feature(bytes_list=tf.train.BytesList(value=[(value)]))

def pickle_to_tfrecord(input_files, output_file):
    print('Processing tfrecord: ' + output_file)
    with tf.python_io.TFRecordWriter(output_file) as record_writer:
        for input_file in input_files:
            path = CIFAR10_PICKLE + '/' + input_file
            print(' Input file : ' + path)
            size, data, labels, filenames = unpickle(path)
            for i in range(size):
                example = tf.train.Example(features=tf.train.Features(
                    feature = {
                        'image': _bytes_feature(data[i].tobytes()),
                        'label': _int64_feature(labels[i])
                    }))
                record_writer.write(example.SerializeToString())

```

# Define input function
* Read training data from tfrecord

```python
# Sepration function of image binary and label from tfrecord

def parse_record(serialized_example):
    features = tf.parse_single_example(
        serialized_example,
        features={
            'image': tf.FixedLenFeature([], tf.string),
            'label': tf.FixedLenFeature([], tf.int64),
        })
    
# array of the image from tfrecord
# chage into one-dimention --> three-dimention(RGB * vertical * horizontal) -> three dimention(vertical * horizontal * RGB)
    image = tf.decode_raw(features['image'], tf.uint8)
    image.set_shape([IMAGE_DEPTH * IMAGE_HEIGHT * IMAGE_WIDTH])
    print('Image.shape 1 : ' + str(image.shape))
    image = tf.reshape(image, [IMAGE_DEPTH, IMAGE_HEIGHT, IMAGE_WIDTH])
    print('Image.shape 2 : ' + str(image.shape))
    image = tf.cast(tf.transpose(image, [1,2,0]), tf.float32)

# read label and apply One Hod Encoding
    label = tf.cast(features['label'], tf.int32)
    label = tf.one_hot(label, NUM_CLASSES)
    print ('Label.shape : ' + str(label.shape))

    return image, label
```

```python
# to prevent overfitting 
# cut the image randomly
# for the improvement of training efficiency, needs various feature engineering

def preprocess_image(image, is_training=False):
    """ Preprocess a single image of layout [height, width, depth]. """
    if is_training:
        # Resize the image to add four extra pixels on each side.
        # make 32 * 32 image to 40 * 40
        image = tf.image.resize_image_with_crop_or_pad(image, IMAGE_HEIGHT + 8, IMAGE_WIDTH + 8)
        
        # Randomly crop a [_HEIGHT, _WIDTH] section of the image.
        image = tf.random_crop(image, [IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_DEPTH])
        
        # Randomly flip the image horizontally.
        image = tf.image.random_flip_left_right(image)
        
        # Subtract off the mean and divide by the variance of the pixels.
        image = tf.image.per_image_standardization(image)
        return image
```

```python
# transfering the trained data

def generate_input_fn(file_names, mode=tf.estimator.ModeKeys.EVAL, batch_size=1):
    def _input_fn():
        # read images from tfrecord file
        # 40,000 for training
        # 10,000 for validation
        dataset = tf.data.TFRecordDataset(filenames=file_names)
        
        is_training = (mode == tf.estimator.ModeKeys.TRAIN)
        if is_training:
            buffer_size = batch_size * 2 + 1
            # Wehn training shuffle the data
            dataset = dataset.shuffle(buffer_size = buffer_size)
            
        # Transformation
        # read data (image, label) in tfrecord file (dataset)
        dataset = dataset.map(parse_record)
        # call function ( preprocess_image) with the 40,000(10,000) data from dataset
        # change the image randomly
        dataset = dataset.map(
            lambda image, label: (preprocess_image(image, is_training), label))
        
        # Keep supply data until whole training is over
        dataset = dataset.repeat()
        dataset = dataset.batch(batch_size)
        dataset = dataset.prefetch(2 * batch_size)
        images, labels = dataset.make_one_shot_iterator().get_next()
        
        # transport to field name images
        features = {'images': images}
        return features, labels
    
    return _input_fn
```

```python
def get_feature_columns():
    feature_columns = {
        'images': tf.feature_column.numeric_column('images', (IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_DEPTH)),
        
    }
    return feature_columns
```

# Create Model

```python
class FLAGS():
    pass

# batch size
FLAGS.batch_size = 200
# Max training numbers
FLAGS.max_steps = 500
# EVAL / TEST 
FLAGS.eval_steps = 100

# Save the reuslt
FLAGS.save_checkpoints_steps = 100
FLAGS.tf_random_seed = 19851211
FLAGS.model_name = 'cnn-model-knu'

# We use a weight decay of 0.0002, which perform better than the 0.0001 that was orginally sugessted
FLAGS.weight_decay = 2e-4
# Momentum Optimizer
FLAGS.momentum = 0.9

# If a model is trained with multiple GPUs, prefix all Op names with tower_name
# to differentiate teh operations. Note that this prefix is removed from the
# names of the summaries wehn visulizing a model.
FLAGS.tower_name = 'tower'
# Continue when it's stopped
FLAGS.use_checkpoint = False
```

```python
def serving_input_fn():
    receiver_tensor = {'images': tf.placeholder(
        shape = [None, IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_DEPTH], dtype=tf.float32)}
    features = {'images': tf.map_fn(preprocess_image, receiver_tensor['images'])}
    return tf.estimator.export.ServingInputReceiver(features, receiver_tensor)
```

```python
# Code define CNN - DNN model

def inference(images, training = False):
    # 64 filter, 5 * 5, relu
    conv1 = tf.layers.conv2d(
        inputs = images,
        filters = 64,
        kernel_size = 5,
        padding = "same",
        activation = tf.nn.relu)
    # Pooling
    pool1 = tf.layers.max_pooling2d(inputs = conv1, pool_size = 3, strides=2)
    # To avoid effect from too much big cost, reduce it to between 0 ~ 1
    norm1 = tf.layers.batch_normalization(pool1, training = training)
    
    # second filter / pooling
    conv2 = tf.layers.conv2d(
        inputs = norm1,
        filters = 64,
        kernel_size = 5,
        padding = "same",
        activation=tf.nn.relu)
    pool2 = tf.layers.max_pooling2d(inputs=conv2, pool_size = 3, strides=2)
    norm2 = tf.layers.batch_normalization(pool2, training=training)
    
    # Change into one-dimention
    # To put into DNN
    norm2_shape = norm2.get_shape()
    dim = norm2_shape[1] * norm2_shape[2] * norm2_shape[3]
    reshape = tf.reshape(norm2, [-1, dim])
    # First hidden lays = 384
    # To precent Vanishing effect, use Relu in hidden layer
    local3 = tf.layers.dense(reshape, 384, activation=tf.nn.relu)

    # drop-out
#    if training:
#        dropout1 = tf.layers.dropout(inputs=local3, rate=0.2, training = mode == tf.estimator.ModeKeys.TRAIN)
#        local4 = tf.layers.dense(dropout1, 192, activation = tf.nn.relu)
    
    # Second hidden layer = 192
    local4 = tf.layers.dense(local3, 192, activation=tf.nn.relu)
    # Output node is 10, as the images is CIFAR-10
    logits = tf.layers.dense(inputs = local4, units = NUM_CLASSES)
    
    return logits
```

```python
# Cacluate the cost

def get_loss(logits, labels):
    # Calculate loss, which includes 'softmax cross entropy and L2 regularization.
    cross_entropy = tf.losses.softmax_cross_entropy(
    logits = logits, onehot_labels=labels)
    
    # Create a tensor named cross_entropy for logging purpose
    tf.identity(cross_entropy, name='cross_entropy')
    tf.summary.scalar('cross_entropy', cross_entropy)
    
    # Add weight decay to the loss.
    loss = cross_entropy + FLAGS.weight_decay * tf.add_n(
        [tf.nn.l2_loss(v) for v in tf.trainable_variables()])
    
    return loss
```

```python
# Specify the parameters required for training
# Only works in Training mode

def get_train_op(loss, params, mode):
    if mode == tf.estimator.ModeKeys.TRAIN:
        # Scale the learning rate linearly with the batch size.
        # When the batch size is 128, the learning rate should be 0.1.
        initial_learning_rate = 0.1 * params.batch_size / 120
        # number of batch per 1 epoch
        batches_per_epoch = NUM_EXAMPLES_PER_EPOCH_FOR_TRAIN / params.batch_size
        # total training number
        global_step = tf.train.get_or_create_global_step()
        
        # Multiply the learninig rate by 0.1 at 100, 150, and 200 epcochs
        boundaries = [int(batches_per_epoch * epoch) for epoch in [100, 150, 200]]
        values = [initial_learning_rate * decay for decay in [1, 0.1, 0.01, 0.001]]
        learning_rate = tf.train.piecewise_constant(
        tf.cast(global_step, tf.int32), boundaries, values)
        
        # Create a tensor named learning_rate for logging purpose
        tf.identity(learning_rate, name='learning_rate')
        tf.summary.scalar('learning_rate', learning_rate)
        
        # Reference : http://shuuki4.github.io/deep%20learning/2016/05/20/Gradient-Descent-Algorithm-Overview.html
        optimizer = tf.train.MomentumOptimizer(
            learning_rate = learning_rate,
            momentum = FLAGS.momentum)
        
        # Batch norm requires update ops to be added as a dependency to the train_op
        update_ops = tf.get_collection(tf.GraphKeys.UPDATE_OPS)
        with tf.control_dependencies(update_ops):
            train_op = optimizer.minimize(loss, global_step)
    else:
        train_op = None
        
    return train_op
```

```python
# Code for calculating accuracy

def get_metrics(predictions, labels):
    accuracy = tf.metrics.accuracy(predictions['classes'], tf.argmax(labels, axis = 1))
    
    # Create a tensor named train_accuracy for logging purposes
    tf.identity(accuracy[1], name='train_accuracy')
    tf.summary.scalar('train_accuracy', accuracy[1])
    
    return {'accuracy : ' : accuracy}
```

```python
# Define whole Tensorflow estimator model
def get_estimator(run_config, hparams):
    def _model_fn(features, labels, mode, params):
        # Crate the input layers from the features
        feature_columns = list(get_feature_columns().values())
        
        images = tf.feature_column.input_layer(
            features = features, feature_columns = feature_columns)
        
        images = tf.reshape(
            images, shape=(-1, IMAGE_HEIGHT, IMAGE_WIDTH, IMAGE_DEPTH))
        
        # Calculate logits through CNN
        if mode == tf.estimator.ModeKeys.TRAIN:
            logits = inference(images, training = True)
        else :
            logits = inference(images, training = False)
            
            
        # Get predictions
        
        predictions = {
            'classes': tf.argmax(logits, axis = 1),
            'probabilities': tf.nn.softmax(logits, name='softmax_tensor')
            
        }
        
        # Provide an estimator spec for 'ModeKeys.PREDICT'
        
        if mode == tf.estimator.ModeKeys.PREDICT:
            export_outputs = {
                'predictions': tf.estimator.export.PredictOutput(predictions)
            }
            return tf.estimator.EstimatorSpec(mode=mode,
                                               predictions = predictions,
                                               export_outputs = export_outputs)
        
        # Define code of calculating cost
        loss = get_loss(logits = logits, labels = labels)
        # Define parameter for training
        train_op = get_train_op(loss = loss, mode = mode, params = params)
        # Define code of calculateing accuracy
        metrics = get_metrics(predictions = predictions, labels = labels)
        
        # Return EstimatorSpec
        return tf.estimator.EstimatorSpec(
            mode = mode,
            predictions = predictions,
            loss = loss,
            train_op = train_op,
            eval_metric_ops = metrics)
    
    return tf.estimator.Estimator(model_fn = _model_fn,
                                     params = hparams,
                                     config = run_config)
```

# Training

```python
model_dir = CIFAR10_MODEL + '/{}' .format(FLAGS.model_name)
print (model_dir)
```

```python
hparams = tf.contrib.training.HParams(
    batch_size = FLAGS.batch_size,
    max_steps = FLAGS.max_steps,
)

# Checkpoint : save the task as a file

run_config = tf.estimator.RunConfig(
    save_checkpoints_steps = FLAGS.save_checkpoints_steps,
    tf_random_seed = FLAGS.tf_random_seed,
    model_dir = model_dir
)

estimator = get_estimator(run_config, hparams)

# There is another Exporter named Final Exporter

exporter = tf.estimator.LatestExporter(
    name = 'Servo',
    serving_input_receiver_fn = serving_input_fn,
    assets_extra = None,
    as_text = False,
    exports_to_keep = 5)

# Define 'how to train' in Estimator model
train_spec = tf.estimator.TrainSpec(
    input_fn = generate_input_fn(
        file_names = [train_data_file],
        mode = tf.estimator.ModeKeys.TRAIN,
        batch_size = hparams.batch_size),
    max_steps = FLAGS.max_steps)

# Define 'how to validate; in Estimator model
eval_spec = tf.estimator.EvalSpec(
    input_fn = generate_input_fn(file_names = [valid_data_file], mode = tf.estimator.ModeKeys.EVAL, batch_size = hparams.batch_size),
    steps = FLAGS.eval_steps,
    exporters = exporter)
```

```python
import shutil

if not FLAGS.use_checkpoint:
    print("Removing previous artifacts...")
    shutil.rmtree(model_dir, ignore_errors = True)
    
# Start training and evaluation
tf.estimator.train_and_evaluate(estimator, train_spec, eval_spec)
```
